```python
#define pessimistic estimate of generalization error function:

def gError( usertree ):
    v1 = 0
    errCount = 0
    while v1 < usertree.tree_.node_count:
        leaf = usertree.tree_.children_left[v1]==-1   
        if leaf == True:
            vIndex = 0
            for val in usertree.tree_.value[v1][0]:
                if vIndex != np.argmax( usertree.tree_.value[v1] ):
                    errCount += usertree.tree_.value[v1][0][vIndex]
                vIndex += 1     
        v1 += 1

    #count leaves
    leaves = 0
    for node in usertree.tree_.children_left:
        if node == -1:
            leaves += 1           

    #total values  
    instances = 0
    for item in usertree.tree_.value[0][0]:
        instances += item

    return errCount + 0.5 * leaves

def getLeaves( usertree ):
    leaves = 0
    for node in usertree.tree_.children_left:
        if node == -1:
            leaves += 1    
    return leaves

def getInstances( usertree ): 
    instances = 0
    for item in usertree.tree_.value[0][0]:
        instances += item
    return instances

def getMisclass( usertree ):
    v1 = 0
    errCount = 0
    while v1 < usertree.tree_.node_count:
        leaf = usertree.tree_.children_left[v1]==-1     
        if leaf == True:
            vIndex = 0
            for val in usertree.tree_.value[v1][0]:
                if vIndex != np.argmax( usertree.tree_.value[v1] ):
                    errCount += usertree.tree_.value[v1][0][vIndex]
                vIndex += 1     
        v1 += 1
    return errCount
```

```python
# 6) Yes, since the left side of the tree includes a split where the prediction on both sides is the same, these leaves could be
# pruned to improve the result. 

# 7)
prunedTree = []
checkedNodes = []
curLeaf = -1
lastLeaf = -1

def dtPruning( usertree, impurityThreshold, start=1 ):
    counter = start
    for node in usertree.tree_.value[start::]:      
        #check current node vs checked
        if counter in checkedNodes:
            counter += 1
            continue
            
        curLeaf = usertree.tree_.children_left[counter]==-1
        if not curLeaf:
            apd = [counter, usertree.tree_.value[counter], curLeaf]
            prunedTree.append( apd )
#***RECURSION            
            dtPruning( usertree, impurityThreshold, counter+1 )

        if curLeaf and prunedTree[-1][2]: #is true
            apd = [counter, node, curLeaf]
            prunedTree.append( apd )
            checkedNodes.append( counter )
            curLeaf = False
            break
        elif curLeaf:
            apd = [counter, node, curLeaf]
            prunedTree.append( apd )
            checkedNodes.append( counter )
            counter += 1
            continue
        elif prunedTree[-1][2] and not prunedTree[-2][2]:
            continue
                    
        if np.argmax( prunedTree[-3][1] ) == np.argmax( prunedTree[-2][1] ) == np.argmax( prunedTree[-1][1] ):
            prunedTree.pop()
            prunedTree.pop()
            prunedTree[-1][2] = True 
            checkedNodes.append( prunedTree[-1][0] )
            curLeaf = False
            continue 
        
        apd = [0, usertree.tree_.value[0][0], False]
        prunedTree.insert(0, apd)
        return prunedTree
```
